```python
import numpy as np
import sympy
from qiskit import *
from qiskit import QuantumCircuit
from qiskit.qasm2 import dumps

from itertools import combinations
from qiskit import Aer, QuantumCircuit
from qiskit.quantum_info import Operator
from qiskit.circuit import Parameter,ParameterVector
from qiskit.utils import algorithm_globals
from qiskit.algorithms.optimizers import OptimizerResult

from qiskit.algorithms.optimizers import L_BFGS_B, COBYLA
from qiskit_aer import StatevectorSimulator
from qiskit.quantum_info import Statevector
import qiskit.quantum_info as qi
import random
```

```python
n_qubits = 10
n_edge = 2 # smaller than n_qibits*(n_qubits-1)/2
qubits = QuantumRegister(n_qubits)
```

```python
def graph_c(qubits,n_edge,edge_chose):
    # edge_chose: list,
    # len(edge_chose)=n_edge, min=0, max= n_qibits*(n_qubits-1)/2-1
    qc = QuantumCircuit(qubits)
    
    for i in qubits:
        qc.h(i)
        
    combinations_lst = list(combinations(qubits, 2))
    for j in range(n_edge):
        edge = combinations_lst[edge_chose[j]]
        qc.cz(edge[0],edge[1])
    return qc
```

```python
edge_list = list(range(int(n_qubits*(n_qubits-1)/2)))
circuit_list = []
for comb in combinations(edge_list, n_edge):
    circuit_list += [graph_c(qubits,n_edge,comb)]
```

```python
intermediate_cost = []
intermediate_index = []

def cost(index, targets, att_cirs):
    '''
    targets: list[dict
            count_dictionary obtained from the probe circuit]
    att_cirs: list[attack circuits], len(att_cirs)=len(targets)-1, because the first
              target refers to empty attack
    '''
    cost = 0
    
    for i in range(len(targets)):
        target = targets[i]
        if i == 0:
            circ = circuit_list[index]
        else:
            cir = circuit_list[index]
            circ = cir.compose(att_cirs[i-1], qubits, inplace=False)
            #circ = cir.compose(att_cirs[i-1], [0,1,2], inplace=False)
            
        result = qi.Statevector.from_instruction(circ).probabilities_dict()
    
        for key, value1 in target.items():
            if key in result.keys():
                value2 = result[key]
            else:
                value2 = 0
            if value1 !=0:
                cost += (value1-value2)**2/value1
    
    intermediate_cost.append(cost)
    intermediate_index.append(index)
    return cost
```
